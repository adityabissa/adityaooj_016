```python
def is_room_clean(room):
  if room == "A":
    return True
  elif room == "B":
    return True
  return False

def clean_room(room):

  print(f"Cleaning room {room}...")
  print(f"cleaned {room}")


if not is_room_clean("A"):
  clean_room("A")


if is_room_clean("A"):
  if not is_room_clean("B"):
    clean_room("B")

  if is_room_clean("B"):
    print("Both rooms A and B are clean!")
```
